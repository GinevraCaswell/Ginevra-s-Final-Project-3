# Project 3: Pesticide exposure alters DNA methylation patterns

### **Scientific Question:** Dose pesticide exposure significantly alter DNA methylation patterns in ovarian tissue of female zebrafish?

In a study by Nathan Donley published in 2019, he investigated over 500 pesticides and found that 13 pesticides approved in the United States of America are banned in at least two or the three of the following nations: EU, Brazil, and China. Of these 13 pesticides, four have been shown to have a positive trend line when looking at usage in pounds per year from 1992 to 2016 (Donley, 2019). The molecular mechanisms in which pesticides affect eukaryotic organisms is essential in determining appropriate agricultural restrictions. 

Epigenetics is the study of surface level chemical tags that alter expression of DNA by making genes more or less accessible to transcriptional activity. Recent research suggests that alterations in epigenetic marks throughout the life of an organism are subject to modification by environmental exposure, including nutritional intake (Zhang, 2012). To investigate DNA methylation at a base-level resolution, whole genome bisulfite (WGBS) sequencing will be used. Significantly altered methylation patterns upon exposure to select pesticides indicate that action should be taken to limit pesticide use in the United States. 

The article by Zhang published in 2012 was found on PubMed. PubMed is a trusted database that provides researchers with current, peer-reviewed publications in the biosciences. 

### **Scientific Hypothesis:** If significant alterations in DNA methylation patterns are seen when comparing results from whole genome bisulfite sequencing (WGBS) of pesticide exposed zebrafish against an unexposed control,  then pesticides exposure significantly alters DNA methylation patterns in key DNA repair pathway genes and therefore they should more carefully restricted in the USA.

A secondary hypothesis for this experiment is that methylation patterns in response to exposure of cells to pesticides are similar enough to indicate that grouped regulation in the US could be implemented. All selected genes have had their fold changes calculated with corresponding q-values. All specified fold change data is statistically significant with q-values < 0.05. 

To test my scientific question and evaluate my hypotheses, I first had to find data that could be used to represent potential data from my experiment. I located the data table of fold change information in the supplementary information of the article entitled "DNA methylation alterations in response to pesticide exposure in vitro." This data was manually inputted.

### **Part 1:** Load the Packages

```python
#Import all required packages
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

### Part 2: Load in the data and perform Bioinformatics Analyses 

Data was hard coded from the supplementary data table 4 from the publication by Zhang, 2012. Genes were selected based on the largest fold change size on either end of the methylation spectrum. This data is fold change data that corresponds to the difference in bisulfite sequencing methylation results before and after exposure to the indicated pesticides.  

In the code below, we can see the methylation fold change data in "fold_change" and the corresponding gene codes in the list "gene_code." The data is then sorted to facilitate representation in the form of a heat plot.  P values are calculated to determine if the genetic expression patterns for each pesticide are significantly similar enough to regulate multiple collectively.

```python
#hard coded data from Suplementary Table 4
gene_code = ['UTF1', 'ARL4C', 'ASPH', 'DZIP1', 'GADD45G', 'EPS8', 'CTGF', 'HTRA1', 'TRHDE', 'IRF8', 'PTPK', 'GATA5', 'LAPTM5', 'CD47', 'PTPN18', 'MANEAL', 'DNAH3', 'RAP1GA1', 'UGT1A6', 'GJB7', 'ZNF23', 'AP3B1']
pesticide = ['fonofos', 'parathion', 'terbufos']
fold_change = [[98.7, 42.1, 38.9, 34.4, 25.2, 23.5, 22.8, 20.7, 18.6, 17.8, 16.9, 15.4, -41.1, -22.8, -9.2, -9.0, -7.4, -6.3, -6.3, -5.9, -5.3, -5.0], 
              [98.0, 114.8, 41.4, 39.5, 23.1, 34.0, 16.4, 37.6, 23.2, 18.5, 14.5, 8.3, -26.6, -21.1, -8.9, -5.9, -4.4, -7.2, -8.1, -4.9, -4.6, -5.7], 
              [201.7, 51.8, 54.3, 22.9, 31.2, 25.1, 27.1, 57.3, 30.0, 21.3, 12.0, 13.6, -36.0, -21.2, -9.4, -9.2, -5.1, -5.4, -7.5, -5.2, -5.2, -5.5]]

#Sort the data so that the values in each list are in accending order
fold_change[0].sort()
fold_change[1].sort()
fold_change[2].sort()

#print organized data
print(fold_change)

#p value comaring fonofos and parathion
pvalue1 = stats.ttest_ind(fold_change[0],fold_change[1])
print(pvalue1)

#p value comaring fonofos and terbufos
pvalue2 = stats.ttest_ind(fold_change[0],fold_change[2])
print(pvalue2)

#p value comaring parathion and terbufos
pvalue3 = stats.ttest_ind(fold_change[1],fold_change[2])
print(pvalue3)
```

### Part 3: Create a heatmap 

Here we use the package seaborn to print out the heatmap shown below. The x-axis will be labeled based on values in the list "gene_code" while the y-axis will be labeled based on values in the list "pesticide." Heatmaps are tools used to display data for analysis. Data values are displayed as relative colors, typically on a scale and this scale has a denoted interpretation. These data matrices can include cluster analyses that visually connect related groups and allow researchers to draw conclusions. The colors in the matrix below correspond to the fold change in CpG methylation, with red representing hyper-methylation (up to about 200 fold change) and blue representing hypo-methylation (down to about -40 fold change).  


```python
#generate NumPy array from fold_change data
heatplot = np.array(fold_change)

#Label y-axis pesticide and x-axis gene_code
x_axis_labels = gene_code
y_axis_labels = pesticide

#adjust figure size
f, ax = plt.subplots(figsize=(17, 8))

#Generate heatmap
map = sns.heatmap(heatplot, annot=True, xticklabels=x_axis_labels, yticklabels=y_axis_labels, ax = ax, cmap = "vlag")
```
